```python
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('corpus')
```

```python
pd.set_option('display.max_columns', None) #to check all the col
df = pd.read_csv('IMDBdata_MainData.csv')
df = df.dropna(subset=['Title', 'Genre', 'Writer', 'Actors', 'Plot', 'Director']) # remove nan rows of interested cols
df = df.reset_index(drop=True) # reset index
df
```

```python
#pre-process
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def clean_text(text):
    text = text.lower() # turn text to lowercase
    text = re.sub('[^a-zA-Z]', ' ', text) # remove non-character
    text = re.sub('\s+', ' ', text) # replace special space with one-space
    return text

def text_tokenize(text):
    text = clean_text(text)
    text_tokenized = nltk.word_tokenize(text) # tokenize word
    last_val = [i for i in text_tokenized if i not in stopwords or len(i)>=3] # remove stopwords and word that its length is lwoer than 3
    return last_val

def sep_val(val):
    # this try-except is for the data that have only 1 element or something we cant split 
    try:
        val = val.split(',')
    except:
        val = val
    last_val = [i.lower().replace(' ', '') for i in val]
    return last_val
```

```python
# df.loc[0,['Title', 'Genre', 'Writer', 'Actors', 'Director', 'Plot']] # just check single row
```

```python
# optional no need to create new
new_df = pd.DataFrame(data=df['Title']) # create new dataframe for cleaned data
new_df
```

```python
new_df['genre'] = df['Genre'].apply(sep_val)
new_df['writer'] = df['Writer'].apply(sep_val)
new_df['actors'] = df['Actors'].apply(sep_val)
new_df['director'] = df['Director'].apply(sep_val)
new_df['plot'] = df['Plot'].apply(text_tokenize)
```

```python
new_df.head()
```

```python
columns = ['genre', 'writer', 'actors', 'director', 'plot']
l = [] # create empty list for cleaned+sum of all input
for i in range(len(new_df)):
    words = ""
    for col in columns:
        words += " ".join(new_df[col][i]) + " "
        # 1st " " is for space between data in col
        # 2nd " " is for space between col
    l.append(words)
```

```python
new_df['cleaned_input'] = l
new_df.head()
```

```python
# feature extraction
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
features = vectorizer.fit_transform(new_df['cleaned_input'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(features, features) # cosine between features and features
```

```python

# it will be something like

#             A          B          C

#    A       1          a          b

#    B       c          1          d

#    C       e          f           1

#    a, b, c, d, e, f --> the similarity between that row and col

#    a diagonal is 1 because row and col is the same
   
#    if we compare it to itself, the similarity will be 1
```

```python
def recommender(title):
    print(f'Movie : {title}')
    print('Recommend : ')
    # find index of title in new_df
    idx = new_df['Title'][new_df['Title'] == title].index[0] # idx -> index of new_df['Title'] where new_df['Title'] == title
    # find the most similar of idx
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    # get index of top10 simil 
    top10 = list(score.iloc[1:11].index) # 1-11 because 0 is itself
#     print(top10)
    
    for i in range(len(top10)):
        # find title in new_df by index
        print(i+1,".", new_df['Title'][top10[i]])
    return " finish ૮ ˶ᵔ ᵕ ᵔ˶ ა "
```

```python
movie_name = 'Harry Potter and the Half-Blood Prince'
recommender(movie_name)
```
