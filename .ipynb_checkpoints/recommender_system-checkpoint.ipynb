```python
import pandas as pd
import numpy as np
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('corpus')
```

```python
# pd.set_option('display.max_columns', None) #to check all the col
df = pd.read_csv('IMDBdata_MainData.csv')
df = df.dropna(subset=['Title', 'Genre', 'Writer', 'Actors', 'Plot', 'Director']) # remove nan rows
df = df.reset_index(drop=True) # reset index
df
```

```python
#pre-process
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
df[['Director']]
```

```python
def clean_text(text):
    text = text.lower() # turn text to lowercase
    text = re.sub('[^a-zA-Z]', ' ', text) # remove non-character
    text = re.sub('\s+', ' ', text) # replace special space with one-space
    return text

def text_tokenize(text):
    text = clean_text(text)
    text_tokenized = nltk.word_tokenize(text) # tokenize word
    last_val = [i for i in text_tokenized if i not in stopwords or len(i)>=3] # remove stopwords and word that its length is lwoer than 3
    return last_val

def sep_val(val):
    try:
        val = val.split(',')
    except:
        val = val
    last_val = [i.lower().replace(' ', '') for i in val]
    return last_val
```

```python
df.loc[0,['Title', 'Genre', 'Writer', 'Actors', 'Director', 'Plot']]
```

```python
new_df = pd.DataFrame(data=df['Title'])
new_df
```

```python
new_df['genre'] = df['Genre'].apply(sep_val)
new_df['writer'] = df['Writer'].apply(sep_val)
new_df['actors'] = df['Actors'].apply(sep_val)
new_df['director'] = df['Director'].apply(sep_val)
new_df['plot'] = df['Plot'].apply(text_tokenize)
```

```python
new_df.head()
```

```python
columns = ['genre', 'writer', 'actors', 'director', 'plot']
l = []
for i in range(len(new_df)):
    words = ""
    for col in columns:
        words += " ".join(new_df[col][i]) + " "
    l.append(words)
```

```python
new_df['cleaned_input'] = l
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
features = vectorizer.fit_transform(new_df['cleaned_input'])
```

```python
from sklearn.decomposition import NMF
from sklearn.preprocessing import Normalizer, MaxAbsScaler
from sklearn.pipeline import make_pipeline
```

```python
scaler = MaxAbsScaler()
nmf = NMF(n_components=50)
normalizer = Normalizer()
pipeline = make_pipeline(scaler,nmf,normalizer)
```

```python
norm_features = pipeline.fit_transform(features)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim = cosine_similarity(features, features)
```

```python
index = pd.Series(new_df['Title'])
index.head()
```

```python
def recommender(title):
    movies = []
    print(f'Movie : {title}')
    print('Recommend : ')
    idx = index[index == title].index[0]
    score = pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top10 = list(score.iloc[1:11].index)
#     print(top10)
    
    for i in top10:
        movies.append(new_df['Title'][i])
    return movies
```

```python
recommender('Harry Potter and the Half-Blood Prince')
```
